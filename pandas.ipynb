```python
netflix = pd.read_csv(pwd + '/netflix_titles.csv')
```

```python
# print(netflix.shape[1])
# print(len(netflix.columns))
# print(netflix.shape[0])
# print(len(netflix))
list(netflix.columns)
```

```python
netflix['type'].unique()
```

```python
netflix[netflix['country'] == 'Singapore'].head()
```

```python
netflix[(netflix['country'] == 'United States') & (netflix['rating'] == 'TV-PG')].head()
```

```python
# Strip leading/trailing spaces in the 'date_added' column
netflix['date_added'] = netflix['date_added'].str.strip()
```

```python
netflix['computer_date'] = pd.to_datetime(netflix['date_added'])

netflix.head()
```

```python
from datetime import datetime as dt
```

```python
netflix['computer_date'].fillna(dt(2020,1,1), inplace = True)
netflix.head()
```

```python
# netflix['computer_date'].dropna()
# netflix[netflix['computer_date'] == '2020-1-1']
```

```python
netflix[['Date Part 1', 'Date Part 2']] = netflix['date_added'].str.split(', ', expand=True)
netflix.head()
```

```python
netflix.rename(columns={'Date Part 2': 'Year'}, inplace=True)
netflix.head()
```

```python
netflix.rename(columns={'Date Part 1': 'Month and Day'}, inplace=True)
netflix.head()
```

```python
netflix['cast'].fillna(value='no cast', inplace=True)
netflix.head()
```

```python
movies_per_country = netflix.groupby('country')['show_id'].count().reset_index()
movies_per_country.head()
```

```python
movies_per_country_in_order = netflix.groupby('country')['show_id'].count().reset_index().sort_values(by='show_id', ascending=False)
movies_per_country_in_order.head()
```

```python
second_dataset = pd.read_csv(pwd + '/netflix_titles_second.csv')

new_dataset = pd.concat([netflix, second_dataset])
new_dataset.head()
```

```python
dataset_merged = pd.merge(left=new_dataset, right=movies_per_country, how='inner', left_on=['country'], right_on=['country'])
dataset_merged.head()
```

```python
dataset_merged.rename(columns={'show_id_y' : 'movies_per_country'}, inplace=True)
dataset_merged.head()
```

```python
pivot_table = netflix.pivot_table(index="country", columns="type", values= "title", aggfunc='count', fill_value=0).reset_index()
pivot_table.head()
```
